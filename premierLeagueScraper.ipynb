```python

import time
from bs4 import BeautifulSoup
from selenium import webdriver
import numpy as np
url = "https://www.premierleague.com/clubs"
driver = webdriver.Chrome(executable_path=r"C:/Users/Kyle/Downloads/chromedriver_win32/chromedriver.exe")
driver.get(url)

soup = BeautifulSoup(driver.page_source, 'lxml')
#print(soup.prettify())

s = soup.find_all('a', class_='indexItem')

for line in s:
    print("This line: /n")
    print(line)
    print("/n")
#     if 'href=' in str(line):  
#         print(line)
```

```python
toSearch = []
for line in s:
    line = str(line)
    idxStart = line.find("clubs")
    idxEnd = line.find(">", idxStart)
    #print(line[idxStart+5:idxEnd])
    teamLinkAddOn = line[idxStart+5:idxEnd-1-8]
    newUrl = url + teamLinkAddOn + 'squad'
    toSearch.append(newUrl)
    #print(teamLinkAddOn)
toSearch
```
